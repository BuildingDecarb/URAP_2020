```python
import pandas as pd

# Load .csv files
emissions_supply = pd.read_csv("Supply-Emissions/emissions-supply.csv")
wh_hourly = pd.read_csv("LoadHourlyProfileData/HP_WH_HourlyProfiles.csv")
sh_hourly = pd.read_csv("LoadHourlyProfileData/HP_SH_HourlyProfiles.csv")

# Generates .csv file with C02 emissions/supply info given a specific region
def es_of_region(region, emissions_supply, wh_hourly, sh_hourly):
    es = emissions_supply[emissions_supply.columns[1]]
    wh_region = wh_hourly[wh_hourly.columns[region]]
    sh_region = sh_hourly[sh_hourly.columns[region]]
    
    emissions_supply = emissions_supply.drop('Unnamed: 0', 1)
    emissions_supply['Region ' + str(region) + ' WH'] = wh_region
    emissions_supply['Region ' + str(region) + ' SH'] = sh_region
    emissions_supply['Region ' + str(region) +  ' WH * C02'] = wh_region * es
    emissions_supply['Region ' + str(region) +  ' SH * C02'] = sh_region * es
    emissions_supply.to_csv('emissions-supply-reg' + str(region) + '.csv')
    
    display(emissions_supply)
```

```python
es_of_region(1, emissions_supply, wh_hourly, sh_hourly)
```
