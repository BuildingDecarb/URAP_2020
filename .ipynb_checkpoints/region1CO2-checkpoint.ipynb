```python
import pandas as pd

emissions_supply = pd.read_csv("Supply-Emissions/emissions-supply.csv")
wh_hourly = pd.read_csv("LoadHourlyProfileData/ER_WH_HourlyProfiles.csv")
sh_hourly = pd.read_csv("LoadHourlyProfileData/HP_SH_HourlyProfiles.csv")

es = emissions_supply[emissions_supply.columns[1]]
wh_region1 = wh_hourly[wh_hourly.columns[1]]
sh_region1 = sh_hourly[sh_hourly.columns[1]]

emissions_supply['Region 1 WH'] = wh_region1
emissions_supply['Region 1 WH * C02'] = wh_region1 * es
emissions_supply['Region 1 SH'] = sh_region1
emissions_supply['Region 1 SH * C02'] = sh_region1 * es

emissions_supply.to_csv('emissions-supply-reg1.csv')
emissions_supply
```
