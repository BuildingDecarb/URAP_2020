```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

year = [1, 32, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335] #first day of month
yearend = [31, 59, 90, 120, 151, 181, 212, 243, 273, 304, 334, 365] #last day of month
lyear = [1, 32, 61, 92, 122, 153, 183, 214, 245, 275, 306, 336] #for leap years
monthnames = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

class Day:
    """
    Class representing a day in the year.
    Attributes:
        weekday (int): day of the week of this Day
        season (string): season of this Day
        use (int[]): hourly energy use
    """
    def __init__(self, use):
        self.season = "winter"
        self.weekday = "weekday"
        self.use = use
        
    def setSeason(self, season):
        self.season = season
    
    def setDay(self, weekday):
        self.weekday = weekday
```

```python
class Year:
    """
    Class representing a year. 
    Attributes:
        name (string): the energy source that the Year is representing
        daylist (Day[]): days in this Year
    """
    def __init__(self, dataframe=None, name=None, column=None):
        self.name = name  
        self.daylist = []  #an array of days
        use = []
        if dataframe is None:
            for i in range(0, 365): #assume 365 days in year
                self.daylist.append(Day(np.zeros(24)))
        else:
            for i in range(1, dataframe.shape[0] + 1):
                use.append(dataframe.loc[i - 1, column])
                if i%24 == 0:
                    self.daylist.append(Day(use))
                    use = []

    def distributeWeekday(self, jan1): 
        """
        Attributes a day of the week for each Day.
        Parameters:
            jan1 (int): day of week for the first day of the year (monday = 1, ...sunday = 7)
        """
        for day in self.daylist:
            if jan1%7 == 6 or jan1%7 == 0:
                day.weekday = 'weekend'
            jan1 = jan1 + 1
    
    def distributeSeason(self):
        """
        Attributes a season {winter, summer} to each Day of the Year, according to PG&E's description.
        """
        i = 1
        for day in self.daylist:
            if i >= year[5] and i < year[9]:
                day.season = 'summer'
                i = i + 1
    
    def sumYear(self):
        """
        Sums daily energy use to find total energy use in the Year.
        """
        yearuse = 0
        for day in self.daylist:
            yearuse = yearuse + sum(day.use)
        return yearuse
```

```python
#The doc that stores information about the total number of customers in each dataset. 
customersumm = pd.read_csv('anonymized_1in10_actual_actual_2014_cluster_summary.csv', index_col=1, header=0)

#The doc that stores energy use data. 
filename = 'sce-res-SCEN-res_misc-noKW-nonCare-0.9_1.0' #change to appropriate filename WITHOUT .csv tag
customers = int(customersumm.loc[filename, 'customer_count'])

sh = "HP_SH_HourlyProfiles.csv"
wh = "HP_WH_HourlyProfiles.csv"

df = pd.read_csv(filename+'.csv')
df = df/customers

#Sets up energy use Year for the base load, with cooling and pool pump use subtracted. 
df['total'] = df['total'] - df['cooling'] - df['poolpump']
df_year = Year(df, "Base Load", 'total')
df_year.distributeSeason()
df_year.distributeWeekday(3)

#Energy use Year for cooling.
cool_year = Year(df, "Cooling", 'cooling')
cool_year.distributeSeason()
cool_year.distributeWeekday(3)

#Energy use Year for pool pump. 
pool_year = Year(df, "Pool Pump", 'poolpump')
pool_year.distributeSeason()
pool_year.distributeWeekday(3)

#Energy use Year for space heating.
df_sh = pd.read_csv(sh)
sh_year = Year(df_sh, "Space Heating", '16')
sh_year.distributeSeason()
sh_year.distributeWeekday(3)

#Energy use Year for water heating.
df_wh = pd.read_csv(wh)
wh_year = Year(df_wh, "Water Heating", '16')
wh_year.distributeSeason()
wh_year.distributeWeekday(3)
```

```python
#Annual energy use by source

def grandSum(categories): 
    """
    Takes an array of Years and sums their dayuse attribute.
    Parameters:
        categories (Year[]): an array of Years of different energy sources
    Returns:
        total (Year): a Year whose dayuse is the sum of every energy source represented in CATEGORIES
    """
    total = Year(name="Total")
    for cat in categories:
        for i in range(0, len(cat.daylist)): #for each day of the year
            for j in range(0, 24): #for each hour of the day
                total.daylist[i].use[j] = total.daylist[i].use[j] + cat.daylist[i].use[j]
    return total

categories = [cool_year, pool_year, df_year, sh_year, wh_year]
total_year = grandSum(categories)

print("Annual energy use from cooling: %f kWh" % cool_year.sumYear())
print("Annual energy use from pool pump: %f kWh" % pool_year.sumYear())
print("Annual energy use from base load: %f kWh" % df_year.sumYear())
print("Annual energy use from space heating: %f kWh" % sh_year.sumYear())
print("Annual energy use from water heating: %f kWh" % wh_year.sumYear())
print("\n")
print("Annual energy use from all sources: %f kWh" % total_year.sumYear())
```

```python
## Graphing residential energy use by source

def graphYear(categories, yeardays):
    """
    Graphs hourly energy use by category by month for each month in the year. Categories are additive.
    Parameters:
        categories (Year[]): array of Years of different energy sources
        yeardays (int[]): the index number of first day of each month, out of the total days in the year 
    """
    plt.figure(figsize = (20,20))
    for i in range(1, len(yeardays) + 1):
        plt.subplot(6, 2, i)
        if i == 12:
            hrAvg = monthAvg(categories, yeardays[i-1], 366)
            graphMonth(categories, hrAvg)
        else: 
            hrAvg = monthAvg(categories, yeardays[i-1], yeardays[i])
            graphMonth(categories, hrAvg)
        plt.grid(True)
        plt.title(monthnames[i - 1])
    plt.show()

def graphMonth(categories, categoryHrAvg):
    """
    Helper function to graph hourly energy use by category in a month. 
    """
    x = range(1, 25)
    runningSum = 0
    for i in range(0, len(categories)):
        plt.plot(x, runningSum + categoryHrAvg[i], label = categories[i].name)
        runningSum = runningSum + categoryHrAvg[i]
    plt.legend()
            
def monthAvg(categories, monthBeg, monthEnd):
    """
    Helper function to average energy use by hour. 
    """
    categoryHrAvg = []
    for x in categories:
        month = x.daylist[monthBeg:monthEnd]
        hourlyAvg = np.zeros(24)
        for i in range(0, len(month)):
            for j in range(0, 24):
                hourlyAvg[j] = hourlyAvg[j] + month[i].use[j]
        hourlyAvg = hourlyAvg/len(month)
        categoryHrAvg.append(hourlyAvg)
    return categoryHrAvg

graphYear(categories, year)
```

```python
## Calculating energy cost for different price plans

def flat(year, price=0.19):
    totaluse = 0
    for day in year.daylist:
        totaluse = totaluse + sum(day.use)
    return totaluse * price


def tier(yearl, baseline=15, tier1=0.22376, tier2=0.28159, tier3=0.49334):
    # https://www.pge.com/tariffs/assets/pdf/tariffbook/ELEC_SCHEDS_E-1.pdf
    monthlycost = []
    i = 1
    totaluse = 0
    for day in yearl.daylist:
        totaluse = totaluse + sum(day.use)
        if i in yearend:
            monthlycost.append(min(baseline*30, totaluse) * tier1 + max(0, totaluse - baseline*30) * tier2 + max(0, totaluse - 4*baseline*30) * tier3)
            totaluse = 0
        i = i + 1
    totalcost = sum(monthlycost)
    return totalcost, monthlycost


def tou(year, speak=0.25354, soffpeak=0.20657, wpeak=0.18022, woffpeak=0.17133):
    # https://www.pge.com/tariffs/assets/pdf/tariffbook/ELEC_SCHEDS_EL-TOU.pdf
    speaksum = 0
    soffpeaksum = 0
    wpeaksum = 0
    woffpeaksum = 0
    monthly1 = []
    monthly2 = []
    monthly3 = []
    monthly4 = []
    i = 1
    for day in year.daylist:
        if day.season == "summer":
            if day.weekday == "weekday":
                speaksum = speaksum + sum(day.use[14:19])
                soffpeaksum = soffpeaksum + sum(day.use[:14]) + sum(day.use[19:])
            else:
                soffpeaksum = soffpeaksum + sum(day.use)
        else:
            if day.weekday == "weekday":
                wpeaksum = wpeaksum + sum(day.use[14:19])
                woffpeaksum = woffpeaksum + sum(day.use[:14]) + sum(day.use[19:])
            else:
                woffpeaksum = woffpeaksum + sum(day.use)
        if i in yearend:
            monthly1.append(speaksum)
            monthly2.append(soffpeaksum)
            monthly3.append(wpeaksum)
            monthly4.append(woffpeaksum)
            speaksum = 0
            soffpeaksum = 0
            wpeaksum = 0
            woffpeaksum = 0
        i = i + 1
        monthlycost = processtou([monthly1, monthly2, monthly3, monthly4], [speak, soffpeak, wpeak, woffpeak])
    return sum(monthlycost), monthlycost
                                 
def processtou(monthly, prices):
    """
    Helper function to calculate energy cost by month for a Year of energy use, following the time of use pricing scheme.
    Parameters:
        monthly (int[][]): [speak use, soffpeak use, wpeak use, woffpeak use]x12 for each month
        prices (int[]): [speak price, soffpeak price, wpeak price, woffpeak price]
    Returns:
        monthlycost (int[]): energy cost by month for a TOU pricing scheme
    """
    monthlycost = []
    for i in range(0, len(monthly[0])):
        monthlycost.append(monthly[0][i] * prices[0] + monthly[1][i] * prices[1] + monthly[2][i] * prices[2] + monthly[3][i] * prices[3])
    return monthlycost
```

```python
print("Yearly energy costs under flat rate plan: $%f" % flat(total_year))
print("Yearly energy costs under tiered rate plan: $%f" % tier(total_year)[0])
print("Yearly energy costs under time of use rate plan: $%f" % tou(total_year)[0])
print("\n")

ppkwh = min(flat(total_year), tier(total_year)[0], tou(total_year)[0])/total_year.sumYear()
print("$/kWh under the cheapest plan: $%f" % ppkwh)
```

```python
no_heating = grandSum([cool_year, pool_year, df_year])

def graphByCat(categories):
    plt.figure(figsize = (20,11)) 
    plt.subplot(2, 2, 1)
    graphtier(categories)
    plt.title("Tiered Plan")
    
    plt.subplot(2, 2, 2)
    graphtou(categories)
    plt.title("Time of Use Plan")
    
    plt.subplot(2, 2, 3)
    graphflat([no_heating])
    graphtier([no_heating])
    graphtou([no_heating])
    plt.title("Total energy use from non-heating sources")
    
    plt.subplot(2, 2, 4)
    graphflat([total_year])
    graphtier([total_year])
    graphtou([total_year])
    plt.title("Total energy use from all sources")
    
    plt.show()

def graphflat(categories):
    x = range(1, 13)
    runningMonthSum = np.zeros(12)
    for cat in categories:
        flatcost = np.repeat(flat(cat)/12,12)
        for i in range(0, 12):
            runningMonthSum[i] = runningMonthSum[i] + flatcost[i]
        plt.plot(x, runningMonthSum, label = cat.name + ", flat")
    plt.legend()
    plt.grid(True)
    plt.xlabel("Month")
    plt.ylabel("$")

def graphtier(categories):
    x = range(1, 13)
    runningMonthSum = np.zeros(12)
    for cat in categories:
        tiercost = tier(cat)[1]
        for i in range(0, 12):
            runningMonthSum[i] = runningMonthSum[i] + tiercost[i]
        plt.plot(x, runningMonthSum, label = cat.name + ", tiered")
    plt.legend()
    plt.grid(True)
    plt.xlabel("Month")
    plt.ylabel("$")

def graphtou(categories):
    x = range(1, 13)
    runningMonthSum = np.zeros(12)
    for cat in categories:
        toucost = tou(cat)[1]
        for i in range(0, 12):
            runningMonthSum[i] = runningMonthSum[i] + toucost[i]
        plt.plot(x, runningMonthSum, label = cat.name + ", tou")
    plt.legend()
    plt.grid(True)
    plt.xlabel("Month")
    plt.ylabel("$")

graphByCat(categories)
```
