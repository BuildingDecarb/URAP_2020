```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
FRE_raw = pd.read_csv("FRE2016_Dan_Kelvin.csv")
TUL_raw = pd.read_csv("TUL2016_Dan_kelvin.csv")
```

Tasks:
- Average kWh/ft^2 of NG & Electric for each building type
- Total square foot of each building type

```python
def avg_statistic_per_bldgtype(statistic, ylabel):
    from matplotlib.pyplot import figure
    avg_datas = np.array([], dtype=np.float64)


    for i in range(TUL_raw['TYPE'].unique().size):
        bldgtype = TUL_raw['TYPE'].unique()[i]
        avg_data = TUL_raw[TUL_raw['TYPE']==bldgtype][statistic].mean(skipna=True)
        avg_datas = np.append(avg_datas, avg_data)

    figure(figsize=(18, 6))
    plt.xlabel("Building Type")
    plt.ylabel(ylabel)
    plt.bar(TUL_raw['TYPE'].unique(), avg_datas)

avg_statistic_per_bldgtype('BLDGAREA', 'Building Area (sq ft)')
```
